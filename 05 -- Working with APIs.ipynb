If you use Twitter API beyond this class, you should obtain your API key here: https://developer.twitter.com/en/apps

````python
from twitter.api import Api
with open('twitter_keys.txt','r') as f:
    CONSUMER_KEY, CONSUMER_SECRET, ACCESS_TOKEN, ACCESS_TOKEN_SECRET = [l.strip() for l in f.readlines()]
VERSION = '20180605'

api = Api(CONSUMER_KEY, CONSUMER_SECRET, ACCESS_TOKEN, ACCESS_TOKEN_SECRET)

l = api.GetSearch(term = '@BuenaVistaUniv', count = 10, since_id = 0)
````

````python
tweets_list = [json.loads(str(t)) for t in l]
df_tweets = pd.DataFrame(tweets_list)
df_tweets[['created_at','id','text']]
````

````python
def collect_tweets(tag, count):
    id = 0
    tweets = []
    while count > 0:
        stream = api.GetSearch(term=tag,count=100,since_id=id)
        for line in stream:
            print(line)
            l = json.loads(str(line))
            tweets.append(l['text'])
            id = l['id']
        count = count - 100
    return tweets

t = collect_tweets('Brexit',500)
````

## GeoPy

Provide lat/lon coordinates based on an address

````python
from geopy.geocoders import Nominatim

def get_coords(inputAddress):
    geolocator = Nominatim(user_agent = 'my-application')
    location = geolocator.geocode(inputAddress)
    return location.latitude,location.longitude

get_coords('The Bit Pit')
````

### Pulling venues data within certain distance from coordinates

````python
radius = 1600 # 1 mile in meters
LIMIT = 100 # max number of venues
latitude, longitude = 42.671843, -95.303868
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    latitude, 
    longitude, 
    radius, 
    LIMIT)
````

````python
results = requests.get(url).json() # Pull and convert to json
````

````python
venues = results['response']['groups'][0]['items']
venues
````

````python
venues[0]
````

````python
from pandas.io.json import json_normalize # transform json to a DataFrame

results = requests.get(url).json()
venues = results['response']['groups'][0]['items']
nearby_venues = json_normalize(venues) # flatten JSON
````

### Parsing venue.categories

Apply method applies the function to each element in the dataframe
```python
apply(<lambda function>) 
```

````python
nearby_venues['venue.categories'][0]
````

````python
# let's make a function to extract the name of the category:
nearby_venues['venues_types'] = nearby_venues['venue.categories'].apply(lambda x:x[0]['shortName'])
nearby_venues
````

````python
# See the 'profile' of the place
nearby_venues.groupby('venues_types').count()
````
